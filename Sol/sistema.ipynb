```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Imagem original
img = cv2.imread("img_cnh.jpg")

res = pytesseract.image_to_string(img)
print(res)

plt.imshow(img)
plt.show()
```

```python
#Converter a imagem para cinza
img_gray = cv2.cvtColor(
    img, 
    cv2.COLOR_BGR2GRAY
)

cv2.imshow('Imagem cinza', img)
cv2.imshow('Imagem colorida',img_gray)        
cv2.waitKey(0)
cv2.destroyAllWindows()

res2 = pytesseract.image_to_string(img_gray)
print(res2)
```

```python
#Limiarização
et,th1 = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)

res3 = pytesseract.image_to_string(th1)
print(res3)

plt.imshow(th1)
plt.show()
```

```python
#Adaptive Thresh Mean
img_gray *= 255
img_gray = np.uint8(img_gray)
adap = cv2.adaptiveThreshold(img_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)

res4 = pytesseract.image_to_string(adap)
print(res4)

plt.imshow(adap)
plt.show()
```

```python
# Adaptive Thresh Gaussian
adapg = cv2.adaptiveThreshold(img_gray,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)

res5 = pytesseract.image_to_string(adapg)
print(res5)

plt.imshow(adapg)
plt.show()

cv2.imshow('Imagem adap', adap)
cv2.imshow('Imagem th1',th1)   
cv2.imshow('Imagem adapg', adapg)     
cv2.waitKey(0)
cv2.destroyAllWindows()
```
