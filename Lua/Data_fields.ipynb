# Junção dos Fields


### IMPORTS

```python
import pandas as pd
import numpy as np
import os
```

### Datasets dos Fields e Soil

```python
field_dict = {}
for csv in os.listdir(path='./Data_fields/'):
    if csv[:5] == "field":
        if csv[7] == '.':
            field_dict[csv[6]] = pd.read_csv(filepath_or_buffer=f'./Data_fields/{csv}')
        else:
            field_dict[csv[6:8]] = pd.read_csv(filepath_or_buffer=f'./Data_fields/{csv}')
    else:
        soil = pd.read_csv(filepath_or_buffer=f'Data_fields/{csv}')
```

```python
for i in range(soil.shape[0]):
    field_dict[str(i)] = pd.concat([soil.loc[i].to_frame().transpose(), field_dict[str(i)]], axis=0)
```

```python
for field in field_dict:
    field_dict[field] = field_dict[field]

```
